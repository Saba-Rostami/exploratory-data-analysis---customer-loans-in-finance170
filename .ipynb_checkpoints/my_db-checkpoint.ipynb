```python
#Â Deleting columns with high percentage of null values 
df.drop(columns=["mths_since_last_delinq","mths_since_last_record","next_payment_date", "mths_since_last_major_derog"], inplace=True)
```

```python
df.rename(columns={'term': 'term_in_month', 
                   'employment_length':"employment_length_in_year"}, inplace=True)
df.head()
```

```python
data_transformer = DataTransform(df)
data_transformer.convert_to_datetime("issue_date")
data_transformer.convert_to_datetime('earliest_credit_line')
data_transformer.convert_to_datetime("last_payment_date")
data_transformer.convert_to_datetime("last_credit_pull_date")
data_transformer.remove_non_digit_charactor("term_in_month")
data_transformer.convert_to_numeric("term_in_month")
data_transformer.remove_non_digit_charactor("employment_length_in_year")
data_transformer.convert_to_numeric("employment_length_in_year")
data_transformer.convert_to_categorical("grade")
data_transformer.convert_to_categorical("sub_grade")
data_transformer.convert_to_categorical("home_ownership")
data_transformer.convert_to_categorical("verification_status")
data_transformer.convert_to_categorical("loan_status")
data_transformer.convert_to_categorical("payment_plan")
data_transformer.convert_to_categorical("purpose")
data_transformer.convert_to_categorical("application_type")


transformed_df = data_transformer.get_transformed_data()
transformed_df.head()
```
